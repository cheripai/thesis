```python
%matplotlib inline
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", color_codes=True)
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
df = pd.read_csv("../data/NDVI/Lettuce_2016-11-03.csv", header=None)
```

```python
df.describe()
```

```python
ax = df.plot(kind="bar", legend=None, title="Lettuce 11/03", color="#3399DD")
for container in ax.containers:
    plt.setp(container, width=0.7)
ax.set_xlabel("Lettuce Plant")
ax.set_ylabel("NDVI")
plt.tight_layout()
```

```python
ax = df.plot(kind="hist", bins=14, color="#3399DD", legend=False)
ax.set_xlabel("NDVI")
```
